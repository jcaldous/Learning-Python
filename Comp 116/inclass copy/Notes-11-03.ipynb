```python
def countD(s):
    r = {}
    for i in range(len(s)):
        si = s[i]
        if si not in r:
            r[si] = 0
        r[si] = r[si] + 1
    return r

countD([1, 2, 3, 42, 19, 3, 2, 1])
```

```python
def countDecorated(s):
    r = {}
    for i in range(len(s)):
        print('i=', i)
        si = s[i]
        print('si=', si)
        if si not in r:
            r[si] = 0
            print('not in', r)
        r[si] = r[si] + 1
        print('r=', r)
    return r

countDecorated([1, 2, 3, 42, 19, 3, 2, 1])
```

```python
counts = countDecorated(['now', 'is', 'the', 'time', 'for', 'all', 'good', 'men', 'to', 'come', 
                'to', 'the', 'aid', 'of', 'their', 'country'])
```
