```python
def MaxRunLength(P):
    result = 0
    N = 0
    for i in range(len(P)):
        if P[i]:
            N = N + 1
        else:
            N = 0
        if N > result:
            result = N
    return result
```

```python
#his solution to f5
def F5(A):
    isDry = A == 0
    counts = []
    for i in range(A.shape[1]): # for each column
        counts.append(MaxRunLength(isDry[:,i]))
    return np.argmax(counts)    
F5(rain)
```
