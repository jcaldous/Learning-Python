```python
def countD(s):
    r={}
    for i in range(len(s)):
        si=s[i]
        if si not in r:
            r[si]=0
        r[si]=r[si] +1
    return r
countD([1,2,3,42,10,3,2,1])
```

```python
counts=countD(['now','is','the','time','for', 'all','good','man', 
        'to', 'come', 'to', 'the', 'aid', 'of', 'their','country' ])
```
